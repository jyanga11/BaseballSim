## Some Starting Definitions

Because "Game Theory" has a pretty big public image in the non-math world, it's important that we start out with clear definitions:

- **Total Conflict**: A game where the final outcome always sums to a *constant* and players gain points only when other players lose them.
  - Another good term might be "Total Adversarial" as that gets across the right idea.

<br>
<br>

- **Zero-Sum**: A special case of a total conflict game where the constant is zero.  
  - Zero-Sum games have a more nuanced and technical interpretation than colloquially understood: it's more than "my loss is your gain"
  - All total conflict games can be rescaled to be "zero-sum" games, so the terms are somewhat redundant
  
<br>
<br>

- **Partial Conflict**: Any game that is not in "Total Conflict".  Players can gain points without other players necessarily losing them.

<br>
<br>

- **Strategy**: A course of action available to a player.

<br>
<br>

- **Outcome**: The end of a game.

<br>
<br>

- **Payoffs**: The results awarded to each player after winning or losing a game.

<br>
<br>

- **Dominant Strategy**: One strategy is dominated by another if every outcome in it is *at least as good as* the corresponding outcome in the other.  

<br>
<br>

- **Dominance Principle**: A rational player should never play a strategy that is dominated by another, e.g., you should never play a "worse" strategy.

<br>
<br>

- **Nash Equilibrium**: An outcome where neither player can benefit by departing *unilaterally* from the strategy associated with that outcome.
  - Named after the famous mathematician John Nash.  Was the subject of his PhD thesis, only a document 28 pages long.

#### A Classic Example

Imagine we have two criminals that are being investigated for a serious crime.  They are being interviewed separately.  Before conducting the interview, each prisoner is told:

- If both criminals remain silent, and admit nothing, they will each serve 1 year in prison
- If only one criminal confesses, then they will serve no time and the other criminal will serve 3 years in prison
- If both criminals confess, they will each serve 2 years in prison



We can make this into a table as follows:

| Criminal 1 Actions | Criminal 2 Actions |         |
|--------------------|--------------------|---------|
|                    |     Stay Silent    | Confess |
|        Stay Silent |        (1,1)       |  (3,0)  |
|            Confess |        (0,3)       |  (2,2)  |

<br>
<br>

What should the prisoners do?

<br>
<br>

---

<br>
<br>
<br>
<br>

- This chart is called a "movement diagram" and details the changes in strategy that benefit the player

- Each person can unilaterally make a better move by confessing

- The "Both Confess" is a stable Nash Equilibrium for the game
    - No individual player can (on their own) make a better move
    - Also called a "Pure Strategy Nash Equilibrium

#### Another Example

Assume we have a baseball pitcher and batter.  The pitcher has to decide to throw either a **fast ball** or a **curve ball** for the batter.  The batter, similarily, must make a guess what the pitcher will throw, and his preparation will allow him to hit with different averages.

Let's assume the following:

<br>

- if the batter guesses a **fast ball** they can hit
  - 0.400 if it IS a fast ball
  - 0.200 if it IS a curve ball

<br>

- if the batter guesses a **curve ball** they can hit
  - 0.100 if it IS a fast ball
  - 0.300 if it IS a curve ball

<br>
<br>

We can make this into a table as follows:

| Batter Guesses | Pitcher Actions |            |
|----------------|-----------------|------------|
|                | Fast Ball       | Curve Ball |
| Fast Ball      | 0.400           | 0.200      |
|   Curve Ball   | 0.100           |    0.300   |

<br>
<br>

The goals of the pitcher and batter are in opposition with one another:

- The pitcher's goal is to *minimize* the batting average of the batter.
- The batter's goal is to *maximize* the batting average

<br>
<br>

What should the batter and pitcher do?  Does a stable Nash equilbrium exist?

<br>
- Total Conflict: improving one player's outcome causes the other player's outcome to worsen

- No Nash Equilibrium exists

- Pure Strategy Solution does not exist

- Requires a **Mixed Strategy**: A randomization of a players pure strategies. Each player executes their pure strategies with a particular probability.

<br>

$q$: probability of anticipating a curveball

    - $q = 1$: Always predicts curveball
    
    - $q = 0$: Always predicts fastball
    
Linear interopolations (Represents movements along the movement diagram):

Batter: $ B \leq 0.1q + 0.4(1-q)$

Pitcher: $ B \leq 0.3q + 0.1(1-q)$

$ 0 \leq q \leq 1$

Maximize these at the SIMULTANEOUSLY!
